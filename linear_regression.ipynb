```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Model:
    def __init__(self):
        self.b1 = np.random.randn()
        self.b0 = np.random.randn()

    def forward(self, x):
        y = self.b0 + self.b1 * x
        return y
    
    def fit(self, X_train, y_train):
        sd_y = y_train.std()
        sd_x = X_train.std()
        mean_y = y_train.mean()
        mean_x = X_train.mean()

        r_xy = np.corrcoef(X_train, y_train)[0, 1]

        self.b1 = (sd_y / sd_x) * r_xy
        self.b0 = mean_y - self.b1 * mean_x

model = Model()
```

```python
xs = [np.random.randn() for x in range(0, 100)]
ys = [x + np.random.randn() for x in xs]

plt.scatter(xs, ys)
```

```python
X_train = [np.random.randn() for x in range(0, 100)]
y_train = [-x + np.random.randn() for x in X_train]

plt.scatter(X_train, y_train)
```

```python
model.fit(np.array(X_train), np.array(y_train))
```

```python
xs = [x for x in range(0, 10)]
ys = [model.forward(x) for x in xs]

plt.plot(xs, ys, c='blue')
plt.scatter(X_train, y_train, c='red')
plt.show()
```
